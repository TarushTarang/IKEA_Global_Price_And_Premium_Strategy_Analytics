```python
exchange_rates = {
    'INR': 1.00,
    'JPY': 0.58,
    'SEK': 10.06,
    'AUD': 64.47,
    'USD': 90.93,
    'EUR': 107.18,    
    'GBP': 122.86 
}

df["price_inr"] = df.apply(lambda row: row["price"] * exchange_rates.get(row["currency"], 1), axis = 1)
```

```python
df.head()
```

```python
df['discount'] = df['discount'].astype(str)
df['discount'] = df['discount'].str.replace('%', '')
df['discount'] = df['discount'].str.replace('Discount ', '')
df['discount'] = pd.to_numeric(df['discount'], errors='coerce')
df['discount'] = df['discount'].fillna(0)
```

```python
df['discount'].unique()[:10]
```

```python
df['discounted_price_inr'] = df['price_inr'] * (1 - df['discount']/100)
```

```python
df.head()
```

```python
df['product_rating'] = pd.to_numeric(df['product_rating'], errors='coerce')
```

```python
df['product_rating_count'] = pd.to_numeric(df['product_rating_count'], errors='coerce')
```

```python
# If product rating count is less then 50,low reliability product.
df['Active Participation'] = df['product_rating_count'] > 50
```

```python
df.isnull().sum()
```

```python
df['has_reviews'] = df['product_rating_count'].fillna(0) > 0
```

df.to_csv("ikea_cleaned.csv", index=False)

```python
df = pd.read_csv("ikea_cleaned.csv")
```

```python
df['price_tier'] = pd.qcut(df['price_inr'], q=3, labels=['Budget', 'Mid-Range', 'Premium'])
```

```python
df.head()
```

```python
def discount_level(d):
    if d == 0:
        return "No Discount"
    elif d <= 10:
        return "Low"
    elif d <= 30:
        return "Medium"
    else:
        return "High"
df['discount_level'] = df['discount'].apply(discount_level)
```

```python
df['discount_level'].unique()
```

```python
def engagement_level(x):
    if x == 0:
        return "No Reviews"
    elif x < 30:
        return "Low"
    elif x < 100:
        return "Medium"
    else:
        return "High"
df['engagement_level'] = df['product_rating_count'].apply(engagement_level)
```
